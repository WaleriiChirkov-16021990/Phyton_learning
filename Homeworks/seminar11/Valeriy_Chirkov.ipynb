```python
# Импорты библиотек и создание функции f(x)
from sympy.plotting import plot
import sympy
x = sympy.Symbol('x')
f_x = - 12 * x ** 4 * sympy.sin(sympy.cos(x)) - 18 * x **3 + 5 * x **2 + 10 * x - 30
```

```python
# Находим производную f(x) и строим ее график.
f_diff_x = f_x.diff()
diff_plot = plot(f_diff_x)
```

```python
# Cтроим график f(x)
plots_ = plot(f_x)
```

```python
# Нажождение приблизительных точек пересечения f(x) с осью "X", для дальнейшего уточнения координат "x", то есть "корней", интересующей нас функции f(x).\
# ( расчет на 2.9 сек.)
res = [i/100 for i in range(-1000, 1000, 1) if f_x.subs(x, i/100) * f_x.subs(x, (i - 1)/100) < 0]
print(res)
```

```python
# Поиск корней f(x) методом подстановки значений \
# с перебором в заданном диапазоне.
# расчет 0.1 сек
for i in range(2273056841, 2273056859):
    g = i/ 1000000000
    if abs(f_x.subs(x, g)) < 0.00000009:
        print(g, f_x.subs(x, g))
```

```python
# подставил найденные корни в функции f(x) и ее производную.
# расчет 0.4 сек
f_x_item = [-7.65062228513, -5.0268659282, -1.33896664, 2.273056846, 4.3835237, 8.0351641334]
result_f_x = []
result_f_diff_x = []
for i in f_x_item:
    result_f_x.append(f_x.subs(x, i))
    result_f_diff_x.append(f_diff_x.subs(x, i))
print(result_f_x)
print(result_f_diff_x)
```

```python
# Нажождение приблизительных точек пересечения производной от f(x) и оси "X", для дальнейшего уточнения координат "x" интересующих меня вершин функции f(x).\
# ( расчет на 33.3 сек.)
res = [i/1000 for i in range(-10000, 10000, 1) if f_diff_x.subs(x, i/1000) * f_diff_x.subs(x, (i - 1)/1000) < 0]
print(res)
```

```python
# Ищу точные координаты вершин, использую производную f(x)
# расчет 0.1 сек
for i in range(700103165851, 700103165861):
    a = i / 100000000000
    if abs(f_diff_x.subs(x, a)) < 0.0000001:
       print(a, f_diff_x.subs(x, a))
```
